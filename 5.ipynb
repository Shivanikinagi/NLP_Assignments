```python
from nltk.corpus import wordnet as wn

def get_wordnet_relations(word):
    print(f"\nWord: {word}")
    print("-" * 40)

    synsets = wn.synsets(word)

    if not synsets:
        print("No entries found in WordNet.")
        return

    # Take the most common meaning
    syn = synsets[0]

    # -----------------------
    # Synonyms
    # -----------------------
    synonyms = set()
    for lemma in syn.lemmas():
        synonyms.add(lemma.name())

    # -----------------------
    # Antonyms
    # -----------------------
    antonyms = set()
    for lemma in syn.lemmas():
        if lemma.antonyms():
            antonyms.add(lemma.antonyms()[0].name())

    # -----------------------
    # Hypernyms (broader terms)
    # -----------------------
    hypernyms = set()
    for hyper in syn.hypernyms():
        hypernyms.add(hyper.name().split('.')[0])

    print("Synonyms :", synonyms)
    print("Antonyms :", antonyms if antonyms else "None")
    print("Hypernyms:", hypernyms)


# -----------------------
# Example usage
# -----------------------
words = ["happy", "car", "teacher", "beautiful"]

for w in words:
    get_wordnet_relations(w)
```

```python
text = "The happy teacher drove a car to the school"

words = text.lower().split()

for word in words:
    get_wordnet_relations(word)
```
