```python
data = {
    "text": [
        "यह फिल्म बहुत अच्छी है",
        "मुझे यह फिल्म पसंद आई",
        "यह मोबाइल बहुत अच्छा है",
        "आज का मौसम बहुत सुहावना है",
        "यह गाना बहुत अच्छा है",
        "मुझे यह ऐप बहुत पसंद है",
        "यह सेवा बहुत अच्छी है",
        "यह खाना बहुत स्वादिष्ट है",

        "मुझे यह फिल्म पसंद नहीं आई",
        "यह मोबाइल बहुत खराब है",
        "यह सेवा बहुत खराब है",
        "मुझे यह गाना पसंद नहीं आया",
        "यह ऐप बहुत खराब है",
        "यह खाना अच्छा नहीं है",
        "मुझे यह फोन पसंद नहीं है",
        "यह अनुभव बहुत खराब था"
    ],
    "sentiment": [
        "positive","positive","positive","positive","positive","positive","positive","positive",
        "negative","negative","negative","negative","negative","negative","negative","negative"
    ]
}
```

```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.naive_bayes import MultinomialNB


df = pd.DataFrame(data)

# Convert text to TF-IDF
vectorizer = TfidfVectorizer()
X = vectorizer.fit_transform(df["text"])
y = df["sentiment"]

# Train model
model = MultinomialNB()
model.fit(X, y)

# Test sentence
test_sentence = ["यह फोन बहुत अच्छा है"]

test_vec = vectorizer.transform(test_sentence)
prediction = model.predict(test_vec)

print("Sentence:", test_sentence[0])
print("Sentiment:", prediction[0])
```
